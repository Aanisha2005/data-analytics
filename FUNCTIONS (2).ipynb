**Q1:What is the difference between a function and a method in Python?**

   ---->Think of **functions** and **methods** like this:

### **Function**  
A function is like a general tool that anyone can use. It doesn’t belong to anyone specifically, and it doesn’t care who uses it. For example:

**Imagine you have a calculator app.**  
- You press a button, type in a number, and it gives you the result.  
- The calculator doesn’t know *who* you are or *where* the numbers come from—it just works when you use it.  

In Python, a function looks like this:
```python
def add(a, b):
    return a + b

print(add(2, 3))  # Output: 5
```

### **Method**  
A method, on the other hand, is like a personal tool that’s part of a specific object. It works only when called on that object, and it can use details about the object itself to do its job.

**Now imagine you have a smartphone.**  
- The "take a photo" feature (a method) works only if you’re using *your* phone. It knows about your camera, your settings, and saves the photo in *your gallery*.  

In Python, a method is tied to an object, like this:
```python
class Phone:
    def __init__(self, owner):
        self.owner = owner

    def take_photo(self):
        return f"{self.owner} is taking a photo!"

my_phone = Phone("Aanisha")
print(my_phone.take_photo())  # Output: Aanisha is taking a photo!
```

---

### The Big Difference:
- **Functions** work anywhere, like a calculator.  
- **Methods** are tied to a specific thing (object) and need it to work, like your phone's camera feature.  




**Q2: Explain the concept of function arguments and parameters in Python.**

  ### **What are Parameters and Arguments?**
When you hear **parameters** and **arguments**, think of **ordering food at a restaurant**.

- **Parameters** are like the menu options. They’re placeholders waiting for specific information.
- **Arguments** are the specific details you provide when you order—like telling the waiter, “I’ll have a large pepperoni pizza.”

In Python:
- **Parameters** are the variables you define in a function.  
- **Arguments** are the actual values you pass to the function when you call it.

---

### **Examples to Make It Clear**

#### **Basic Example:**
```python
# Parameters (placeholders)
def greet(name):
    print(f"Hello, {name}!")

# Arguments (specific value)
greet("Aanisha")  # Output: Hello, Aanisha!
greet("John")    # Output: Hello, John!
```

Here:
- `name` is the parameter.
- `"Aanisha"` and `"John"` are the arguments.

---

### **Types of Function Arguments**

1. **Positional Arguments**  
   These are the most common type. The arguments are matched to the parameters in the same order.

   Example:
   ```python
   def introduce(name, age):
       print(f"My name is {name} and I am {age} years old.")

   introduce("Aanisha", 20)  # Output: My name is Aanisha and I am 20 years old.
   ```

   Just like saying, "I want a **pepperoni pizza, large size**."

2. **Keyword Arguments**  
   These let you specify which parameter gets which argument by name. The order doesn’t matter!

   Example:
   ```python
   introduce(age=20, name="Aanisha")  # Output: My name is Aanisha and I am 20 years old.
   ```

   It's like saying, "I want a pizza with **large size and pepperoni topping**."

3. **Default Parameters**  
   You can set a **default value** for a parameter, so it works even if no argument is given.

   Example:
   ```python
   def greet(name="Guest"):
       print(f"Hello, {name}!")

   greet()           # Output: Hello, Guest!
   greet("Aanisha")   # Output: Hello, Aanisha!
   ```

   Imagine the waiter assumes **"medium size"** if you don’t specify the size.

4. **Variable-Length Arguments**  
   Sometimes, you don’t know how many arguments you’ll need. You can use `*args` (for positional) or `**kwargs` (for keyword arguments).

   Example 1: **`*args` for multiple arguments**
   ```python
   def add_numbers(*numbers):
       total = sum(numbers)
       print(f"The total is: {total}")

   add_numbers(1, 2, 3)       # Output: The total is: 6
   add_numbers(5, 10, 15, 20) # Output: The total is: 50
   ```

   It's like saying, "Add as many toppings as you want!"

   Example 2: **`**kwargs` for keyword arguments**
   ```python
   def display_details(**details):
       for key, value in details.items():
           print(f"{key}: {value}")

   display_details(name="Aanisha", age=20, city="Chennai")
   # Output:
   # name: Aanisha
   # age: 20
   # city: Chennai
   ```

   This is like saying, "Here are all the special instructions for my pizza!"

---

### Recap:
- **Parameters**: The placeholders for information (like the menu).
- **Arguments**: The actual information you provide (like your specific order).


**Q3: What are the different ways to define and call a function in Python?**



### **What Does It Mean to Define and Call a Function?**
- **Defining a Function**: It’s like writing the recipe for a dish. You explain *how* to cook it, but nothing happens yet.  
- **Calling a Function**: It’s like actually making the dish by following the recipe. This is when the magic happens!

In Python, you define a function using the `def` keyword, and you call it by using its name followed by parentheses `()`.

---

### **Different Ways to Define and Call a Function**

#### 1. **Simple Function Without Parameters**
- **Definition**: A function that doesn’t take any input.
- **Call**: Just use its name followed by `()`.

**Example**:
```python
# Define the function
def greet():
    print("Hello, world!")

# Call the function
greet()  # Output: Hello, world!
```

---

#### 2. **Function With Parameters**
- **Definition**: A function that takes input (parameters) so you can customize its behavior.
- **Call**: Provide the input values (arguments) inside the parentheses.

**Example**:
```python
# Define the function
def greet(name):
    print(f"Hello, {name}!")

# Call the function with arguments
greet("Aanisha")  # Output: Hello, Aanisha!
greet("John")    # Output: Hello, John!
```

---

#### 3. **Function With a Return Value**
- **Definition**: A function that calculates or processes something and gives you the result (using the `return` keyword).
- **Call**: Store the result in a variable or use it directly.

**Example**:
```python
# Define the function
def add(a, b):
    return a + b

# Call the function and store the result
result = add(5, 3)
print(result)  # Output: 8

# Or use it directly
print(add(10, 20))  # Output: 30
```

---

#### 4. **Function With Default Parameters**
- **Definition**: A function with a default value for a parameter. It’s optional to pass an argument.
- **Call**: Call it with or without an argument.

**Example**:
```python
# Define the function
def greet(name="Guest"):
    print(f"Hello, {name}!")

# Call the function
greet()           # Output: Hello, Guest!
greet("Aanisha")   # Output: Hello, Aanisha!
```

---

#### 5. **Function With Variable-Length Arguments (`*args` and `**kwargs`)**
- **Definition**: A function that can take a flexible number of inputs.
- **Call**: Pass as many arguments as needed.

**Example 1: `*args` for multiple positional arguments**:
```python
# Define the function
def add_all(*numbers):
    total = sum(numbers)
    print(f"The total is: {total}")

# Call the function with varying arguments
add_all(1, 2, 3)       # Output: The total is: 6
add_all(4, 5, 6, 7, 8) # Output: The total is: 30
```

**Example 2: `**kwargs` for multiple keyword arguments**:
```python
# Define the function
def display_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

# Call the function
display_info(name="Anisha", age=20, city="Chennai")
# Output:
# name: Aanisha
# age: 20
# city: Chennai
```

---

#### 6. **Lambda (Anonymous) Function**
- **Definition**: A small, one-line function without a name (used for simple tasks).
- **Call**: Use it directly or assign it to a variable.

**Example**:
```python
# Define and call directly
print((lambda x, y: x + y)(5, 3))  # Output: 8

# Assign to a variable and call
add = lambda x, y: x + y
print(add(10, 20))  # Output: 30
```

---

### **Recap**
- **Define a function**: Write its “recipe” using `def`.
- **Call a function**: Use its name and provide inputs (if needed).  
Different ways:
1. Simple function (no parameters).
2. Function with parameters.
3. Function with a return value.
4. Function with default values.
5. Functions with flexible inputs (`*args`/`**kwargs`).
6. Lambda (short, anonymous) functions.



**Q4: What is the purpose of the `return` statement in a Python function?**

### **What Does `return` Do?**
The `return` statement is like a **delivery guy** bringing something from a function back to you. Without it, the function can only show you the result (like printing it), but you can't use it later.

---

### **Why Use `return`?**
1. **To Give Results**: It sends the answer (output) of the function back to you.
2. **To Reuse Results**: You can store the result in a variable and use it in other tasks.
3. **To End the Function**: It stops the function as soon as it's used.

---

### **Examples of `return`**

#### 1. **Basic Example**
```python
def add(a, b):
    return a + b  # Sends the result back

result = add(5, 3)  # Stores the result
print(result)  # Output: 8
```
Here:
- The function calculates `5 + 3` and *returns* `8`.
- You store it in `result` to use it later.

---

#### 2. **Without `return`**
```python
def add(a, b):
    print(a + b)  # Only shows the result

result = add(5, 3)  # Nothing is returned
print(result)  # Output: None
```
Here:
- The function only **shows** the result with `print`.
- You can’t reuse the result because `return` wasn’t used.

---

#### 3. **Reusing Returned Results**
```python
def square(num):
    return num ** 2

value = square(4) + 10  # Use the result in another calculation
print(value)  # Output: 26
```
Here:
- The `square` function returns `16`, which you add to `10` to get `26`.

---

### **Think of It Like This:**
- **With `return`**: It’s like getting a pizza delivered—you can eat it now or save it for later.
- **Without `return`**: It’s like the pizza shop showing you the pizza but not letting you keep it.

**Q5:What are iterators in Python and how do they differ from iterables?**


### **What Are Iterators and Iterables?**
Think of **iterables** and **iterators** as part of a **conveyor belt system** in a factory:

- **Iterable**: It’s the entire conveyor belt with items on it. You can walk along and pick up one item at a time.  
- **Iterator**: It’s like the person picking up each item. The iterator knows where it is and how to get the next item.

---

### **Key Points**
1. **Iterable**: An object you can loop over (like a list, string, or tuple). It has a method called `__iter__()` that makes it ready for iteration.
2. **Iterator**: An object that keeps track of where it is in the iteration. It has two methods:
   - `__iter__()` (to make it an iterator).
   - `__next__()` (to get the next item in the sequence).

---

### **Difference Between Iterables and Iterators**

| **Feature**        | **Iterable**                             | **Iterator**                          |
|---------------------|------------------------------------------|---------------------------------------|
| **Definition**      | An object you can loop over.             | An object that produces items one at a time. |
| **Methods**         | Has `__iter__()` method.                 | Has `__iter__()` and `__next__()` methods. |
| **Examples**        | Lists, strings, tuples, dictionaries.    | Objects created by calling `iter()` on an iterable. |
| **State Tracking**  | Does not keep track of where you are.    | Keeps track of the current position.  |

---

### **Examples to Understand Iterables and Iterators**

#### 1. **Iterable Example**
```python
# A list is an iterable
my_list = [1, 2, 3, 4]

# You can loop over it
for item in my_list:
    print(item)
# Output:
# 1
# 2
# 3
# 4
```
Here:
- `my_list` is an **iterable** because you can loop over it with a `for` loop.

---

#### 2. **Iterator Example**
```python
# Create an iterator from a list
my_list = [1, 2, 3, 4]
my_iterator = iter(my_list)  # Convert to iterator

# Get items one by one
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
print(next(my_iterator))  # Output: 4
```
Here:
- `my_iterator` is an **iterator**.
- The `next()` function gets the next item in the sequence.
- When the iterator runs out of items, calling `next()` will raise a `StopIteration` error.

---

### **How They Work Together**
An **iterator** is created from an **iterable** using the `iter()` function.  
- For example, calling `iter([1, 2, 3])` turns the list into an iterator.

---

### **Recap:**
- An **iterable** is like a playlist on a music app. It has all your songs.
- An **iterator** is the play button—it plays one song at a time and moves to the next song when you press it.

**Q6:Explain the concept of generators in Python and how they are defined**


### **What Are Generators?**
Generators are like **TV series episodes on-demand**:  
- Instead of downloading all the episodes at once, you stream one episode at a time when you need it.  
This saves memory and time!

In Python, **generators** produce values one at a time instead of creating all the values at once (like a list does).  

---

### **How Are Generators Defined?**
Generators are created using a function with the `yield` keyword.  
- **`yield`** is like a **pause button**: It saves the function’s state and continues from there the next time you call it.

---

### **Difference Between `yield` and `return`**
- **`return`**: Ends the function and sends back one value.
- **`yield`**: Pauses the function, remembers where it left off, and lets you resume later.

---

### **How to Define and Use Generators**

#### 1. **Creating a Simple Generator**
```python
def my_generator():
    yield 1  # Pause and give back 1
    yield 2  # Pause and give back 2
    yield 3  # Pause and give back 3

# Using the generator
gen = my_generator()  # Create the generator
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```
Here:
- `my_generator()` doesn’t give all values at once. It produces one value at a time when you call `next()`.

---

#### 2. **Using Generators in Loops**
Generators work well with loops because they give one value at a time.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Pause and return the current value
        count += 1

# Using the generator
for num in count_up_to(5):
    print(num)
# Output:
# 1
# 2
# 3
# 4
# 5
```
Here:
- The generator gives numbers from `1` to `5` one by one.

---

### **Why Use Generators?**

1. **Memory Efficient**: Instead of creating a huge list in memory, generators only produce values when needed.
   ```python
   def big_range(n):
       for i in range(n):
           yield i

   gen = big_range(1000000)  # Doesn't use up memory
   ```

2. **Infinite Sequences**: Generators can create infinite sequences (like Fibonacci numbers) without running out of memory.
   ```python
   def infinite_numbers():
       num = 1
       while True:
           yield num
           num += 1
   ```

---

### **Recap:**
- A **list** is like a pre-loaded playlist with all songs downloaded.  
- A **generator** is like a streaming app—it plays songs one by one as you listen, without downloading all at once.

---



**Q7:What are the advantages of using generators over regular functions?**

Generators are like **smart delivery services**—they send you one item at a time when you need it, instead of dropping everything at your doorstep at once. This makes them very efficient and flexible!

---

### **Advantages of Generators Over Regular Functions**

1. **Saves Memory**  
   - Regular functions often create entire data structures (like lists) in memory, even if you don’t use all the values.  
   - Generators **produce values one by one**, so they don’t need to store everything at once.  
   
   **Example**:  
   ```python
   def create_list(n):
       return [i for i in range(n)]  # Creates a big list

   def create_generator(n):
       for i in range(n):
           yield i  # Produces one number at a time
   ```
   - If `n = 1,000,000`, the list will take a lot of memory, but the generator won’t!

---

2. **Faster Execution (Lazy Evaluation)**  
   - Generators don’t compute all the values immediately. They **only calculate when you ask for them**, which makes them faster in certain cases.  
   - Regular functions calculate everything upfront, even if you don’t need all the data.

---

3. **Easier to Work With Large or Infinite Data**  
   - Generators are perfect for **large datasets** or **infinite sequences** because they don’t load everything into memory.  
   
   **Example**: An infinite number generator:
   ```python
   def infinite_numbers():
       num = 1
       while True:
           yield num
           num += 1
   ```
   - With a regular function, you’d crash your system trying to create an infinite list!

---

4. **Cleaner and Simpler Code**  
   - Generators can simplify code by avoiding the need for temporary data structures like lists.  
   
   **Example**: Find even numbers from 1 to 10.  
   **With Regular Function**:
   ```python
   def get_evens(n):
       evens = []
       for i in range(1, n + 1):
           if i % 2 == 0:
               evens.append(i)
       return evens
   print(get_evens(10))  # Output: [2, 4, 6, 8, 10]
   ```
   **With Generator**:
   ```python
   def get_evens(n):
       for i in range(1, n + 1):
           if i % 2 == 0:
               yield i
   for num in get_evens(10):
       print(num, end=" ")  # Output: 2 4 6 8 10
   ```

---

5. **Flexible for Iteration**  
   - Generators work seamlessly in loops (`for` loops) because they give one value at a time. No need to pre-generate or store the entire sequence!

---

### **Recap:**
- **Regular Function**: It’s like getting 100 pizzas delivered all at once—even if you only want one slice.  
- **Generator**: It’s like a pizza shop that delivers one slice at a time, as you’re hungry. It’s faster, uses less space, and avoids waste.

**Q8: What is a lambda function in Python and when is it typically used?**


### **What is a Lambda Function?**

A **lambda function** in Python is a **small, anonymous function** that can be written in a single line. It’s like a **quick shortcut** when you need a small function, but you don’t want to go through the trouble of naming it.

It’s also called an **anonymous function** because it doesn’t have a name.

---

### **Syntax of a Lambda Function**  
A lambda function follows this simple structure:
```python
lambda arguments: expression
```

- **`lambda`**: Keyword that defines the function.
- **`arguments`**: The input values to the function (like parameters).
- **`expression`**: The operation or result you want to return.

---

### **When Do You Use Lambda Functions?**
1. **When You Need a Quick Function**  
   If you only need a small function for a short task, like adding two numbers or checking if a number is even, a lambda is perfect!

2. **With Functions like `map()`, `filter()`, and `sorted()`**  
   These functions let you apply the lambda function to each item in a collection (like a list or tuple) to process data quickly.

---

### **Examples of Lambda Functions**

#### 1. **Simple Addition**
```python
# Regular function
def add(x, y):
    return x + y

# Lambda function
add_lambda = lambda x, y: x + y

print(add_lambda(3, 5))  # Output: 8
```
Here:
- `lambda x, y: x + y` is the **lambda function** doing the same thing as the regular `add` function.

---

#### 2. **Using Lambda with `sorted()`**
```python
# List of tuples
students = [("John", 85), ("Anna", 92), ("Mike", 78)]

# Sort by the second value (marks) using lambda
sorted_students = sorted(students, key=lambda student: student[1])

print(sorted_students)  
# Output: [('Mike', 78), ('John', 85), ('Anna', 92)]
```
Here:
- The `lambda student: student[1]` function is used to sort students by their marks.

---

#### 3. **With `filter()` to Check Even Numbers**
```python
numbers = [1, 2, 3, 4, 5, 6]

# Filter even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)  # Output: [2, 4, 6]
```
Here:
- `lambda x: x % 2 == 0` checks if a number is even.

---

### **Recap:**
- A **lambda function** is like a **quick, one-time tool** you use for a small task, instead of creating a whole tool (regular function) for it.
- It’s great for **short tasks** where you don’t need a name or reuse the function.

**Q9:Explain the purpose and usage of the `map()` function in Python.**

### **What is the `map()` Function?**

The **`map()`** function is like a **magic worker** that takes a function and applies it to every item in a list (or any iterable like a list, tuple, etc.). It helps you **transform** the items in that list without needing a loop.

Think of it as a **batch processor**: it works on every item in the list **at once** and gives you back a new list with the results.

---

### **Syntax of `map()`**

```python
map(function, iterable)
```

- **`function`**: The function you want to apply to each item in the iterable.
- **`iterable`**: The collection of items (like a list) you want to process.

---

### **How `map()` Works:**

1. **Takes each item** from the list or collection.
2. **Applies the function** to that item.
3. **Gives back** a new collection with the results.

---

### **Examples of Using `map()`**

#### 1. **Simple Example: Adding 1 to Each Number**
```python
numbers = [1, 2, 3, 4]

# Function to add 1
def add_one(x):
    return x + 1

# Apply add_one to every number in the list
new_numbers = list(map(add_one, numbers))

print(new_numbers)  # Output: [2, 3, 4, 5]
```

Here:
- The `map()` function takes each item from `numbers`, applies `add_one` to it, and returns a new list of numbers with 1 added to each.

---

#### 2. **Using `map()` with Lambda**
You can use **lambda functions** (quick, one-time functions) inside `map()` for short tasks:

```python
numbers = [1, 2, 3, 4]

# Using lambda to add 1 to each number
new_numbers = list(map(lambda x: x + 1, numbers))

print(new_numbers)  # Output: [2, 3, 4, 5]
```

Here:
- The **lambda function** `lambda x: x + 1` is used to add 1 to each number in the list.

---

#### 3. **Using `map()` with Multiple Iterables**
You can also use `map()` with more than one list. It applies the function to the **items from both lists at the same time**.

```python
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

# Add corresponding numbers from both lists
result = list(map(lambda x, y: x + y, numbers1, numbers2))

print(result)  # Output: [5, 7, 9]
```

Here:
- The function adds **corresponding elements** from `numbers1` and `numbers2` (1 + 4, 2 + 5, 3 + 6).

---

### **Recap:**
- **`map()`** is like an **assembly line** where you apply the same operation (like adding or changing something) to every item in a list automatically, without writing a loop.
- It **saves time and effort**, especially if you want to do the same thing to a lot of items in a list!

**Q10: What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**


### **1. `map()` Function:**

- **What It Does**:  
  `map()` helps you **change** every item in a list. For example, you can add 1 to each number in a list.

- **When to Use**:  
  When you want to **apply the same operation** (like adding, multiplying) to each item in a list.

- **Example**:  
  Imagine you have a list of numbers and you want to **add 1 to each**.
  ```python
  numbers = [1, 2, 3]
  result = list(map(lambda x: x + 1, numbers))
  print(result)  # Output: [2, 3, 4]
  ```
  Here:
  - `map()` takes each number from the list and **adds 1** to it.

---

### **2. `reduce()` Function:**

- **What It Does**:  
  `reduce()` helps you **combine** all the items in a list into a **single value**. It's like **mixing everything together** into one result.

- **When to Use**:  
  When you want to **combine** items in a list into one single result, like adding them up or multiplying them.

- **Example**:  
  Let’s say you want to **multiply all numbers** in a list.
  ```python
  from functools import reduce

  numbers = [1, 2, 3, 4]
  result = reduce(lambda x, y: x * y, numbers)
  print(result)  # Output: 24 (1 * 2 * 3 * 4)
  ```
  Here:
  - `reduce()` **multiplies** all numbers together to get a single result (24).

---

### **3. `filter()` Function:**

- **What It Does**:  
  `filter()` helps you **pick out** only the items that **meet a certain condition**. For example, you can **pick only even numbers** from a list.

- **When to Use**:  
  When you want to **filter out** items that don’t meet your condition (like filtering even numbers from a list).

- **Example**:  
  Imagine you have a list of numbers, and you want to **keep only the even numbers**.
  ```python
  numbers = [1, 2, 3, 4, 5, 6]
  result = list(filter(lambda x: x % 2 == 0, numbers))
  print(result)  # Output: [2, 4, 6]
  ```
  Here:
  - `filter()` **keeps only the even numbers** from the list.

---

### **Summary of Differences:**

- **`map()`**: Changes **every item** in a list.  
  **Example**: Add 1 to each number.

- **`reduce()`**: **Combines** all items into one result.  
  **Example**: Multiply all numbers together.

- **`filter()`**: **Keeps items** that match a condition.  
  **Example**: Keep only even numbers.

---

### **In Super Simple Terms:**
- **`map()`**: Imagine you have a list of fruits, and you want to **wash each one** (apply the same action to all).
- **`reduce()`**: Imagine mixing all the fruits into a **smoothie** (combining all items into one).
- **`filter()`**: Imagine you have a basket of fruits, and you only want to **keep the ripe ones** (selecting only the ones that fit a condition).

### **What is `reduce`?**
The `reduce` function, from the `functools` module in Python, reduces a list to a single value by applying a function cumulatively to the items in the list.

- **Cumulatively** means: Start with the first two items, then take the result and combine it with the next item, and so on, until one final result is left.

---



### Steps for Writing:

1. **Given List**: `[47, 11, 42, 13]`
   - The `reduce()` function takes two inputs at a time and performs the operation cumulatively across all elements.

2. **Process**:
   - Step 1: Take the first two elements, `47` and `11`.
     - Sum: `47 + 11 = 58`
   - Step 2: Take the result (`58`) and the next element (`42`).
     - Sum: `58 + 42 = 100`
   - Step 3: Take the result (`100`) and the last element (`13`).
     - Sum: `100 + 13 = 113`

3. **Write it Down Step-by-Step**:
   ```
   reduce(lambda x, y: x + y, [47, 11, 42, 13])
   Step 1: x = 47, y = 11 → 47 + 11 = 58
   Step 2: x = 58, y = 42 → 58 + 42 = 100
   Step 3: x = 100, y = 13 → 100 + 13 = 113
   ```

4. **Final Result**:
   - **113**

````python
#Q4: Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Check for numbers from 1 to 200
primes = [num for num in range(1, 201) if is_prime(num)]
print("Prime numbers from 1 to 200:", primes)
````

````python
#Q5:Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example usage
fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
````

````python
#Q6:Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(5):
    print(power)
````

````python
#Q7:Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    """
    A generator function to read a file line by line.

    Args:
        file_path (str): The path to the file.

    Yields:
        str: A line from the file as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Removes trailing newline characters
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# for line in read_file_line_by_line("example.txt"):
# print(line)



````

````python
#Q9: Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Using map() to convert temperatures
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Output the results
print("Temperatures in Celsius:", celsius_temps)
print("Temperatures in Fahrenheit:", fahrenheit_temps)
````

````python
#Q10:Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_string = "Python is amazing!"

# Using filter() to remove vowels
result = ''.join(filter(is_not_vowel, input_string))

# Output the result
print("Original string:", input_string)
print("String without vowels:", result)
````

````python
#Q11: Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.

#Write a Python program using lambda and map.
# Example list of orders: [Order Number, Price per Item, Quantity]
orders = [
    [34587, 4.50, 4],
    [98762, 10.00, 5],
    [77226, 2.50, 1],
    [88112, 6.00, 8]
]

# Lambda function to calculate the required tuple
process_order = lambda order: (
    order[0],
    order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
)

# Using map() to process the orders
processed_orders = list(map(process_order, orders))

# Output the result
print("Processed Orders:", processed_orders)
````
