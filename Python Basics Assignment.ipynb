```python
#5  What is mutability in Python?
      #Mutability in Python is about whether you can change something after creating it.
      #For example, you can add or remove items from a list like `[1, 2, 3]` (mutable), but you can’t change a string like `"hello"` directly (immutable).
```

```python
#6 Why are lists mutable, but tuples are immutable?
      #Lists are mutable because they’re meant to be flexible, so you can change them, like adding items to a list (`[1, 2, 3]`).
      #Tuples, on the other hand, are immutable because they’re meant to keep their data fixed and unchangeable (`(1, 2, 3)`).
```

```python
#7 What is the difference between “==” and “is” operators in Python?
      #The `==` operator checks if two things have the same value, while `is` checks if they’re actually the same thing in memory.
      #For example, `a == b` checks if the values are the same, but `a is b` makes sure they’re the same object, not just two copies.
```

```python
#8 What are logical operators in Python?
      #Logical operators in Python help combine conditions in your code.
      #For example, `and` checks if both conditions are true (`a > 10 and b < 15`), `or` checks if at least one is true, and `not` reverses a condition (`not a`).
```

```python
#9 What is type casting in Python?
      #Type casting in Python is like changing something from one form to another.
      #For example, you can turn an integer into a string with `str(9)`, or a string into an integer with `int("35")`.
```

```python
#10 What is the difference between implicit and explicit type casting?
      #Implicit type casting happens automatically when Python changes a smaller data type to a bigger one, like turning an integer into a float.
#implicit type casting example
x = 11
y = 11.0
x + y  # x gets automatically converted to float

      #Explicit type casting is when you manually convert one type to another, like using `int()` to turn a float into an integer.
#explicit typecasting example
x = 5.6
int(x)  # Explicitly converting float to integer
```

```python
#11  What is the purpose of conditional statements in Python?
      #Conditional statements in Python let you make decisions in your code based on certain conditions.
      #For example, an `if` statement checks if something is true and runs the code inside if it is:
#'if' statement
x = 8
if x > 10:
    print("x is greater than 10")

#You can also use `else` for when the condition is false
else:
    print("x is not greater than 10")
#These help your program respond differently based on different situations.
```

```python
#12 How does the elif statement work?
     #The `elif` statement lets you check multiple conditions in a row. If the first `if` condition isn’t true, it moves on to `elif` to check another condition, and so on. Here’s an example:
#example of elif statement
x = 15
if x < 10:
    print("x is less than 10")
elif x == 15:
    print("x is 15")
else:
    print("x is greater than 10 but not 15")

#Since `x` is 15, it will print "x is 15" because that matches the `elif` condition. If none of the conditions match, the `else` part will run.
```

```python
#13 What is the difference between for and while loops?
      # The `for` loop is used when you know in advance how many times you want to repeat something.
      #It loops over a sequence like a list or a range of numbers. For example:
#for loop
for i in range(3):
    print(i)
#This will print `0`, `1`, and `2` because it repeats exactly 3 times.
#The `while` loop, on the other hand, keeps running as long as a condition is true. You don’t always know how many times it will repeat. For example:
#while loop
x = 0
while x < 3:
    print(x)
    x += 1
#This also prints `0`, `1`, and `2`, but it continues until the condition (`x < 3`) is no longer true.
```

```python
#14 Describe a scenario where a while loop is more suitable than a for loop?
# Example: Using a while loop to validate user input
number = -1  # Initialize with an invalid value

while number <= 0:  # Loop until the user enters a positive number
    try:
        number = int(input("Enter a positive number: "))
        if number <= 0:
            print("That's not a positive number. Try again.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

print(f"Thank you! You entered {number}.")
```

```python
#14.Write a program to reverse a string using a while loop.
# Input string from the user
original_string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(original_string) - 1  # Start from the last character

# Reverse the string using a while loop
while index >= 0:
    reversed_string += original_string[index]
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_string)
```

```python
#15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
factorial, i = 1, 1
while i <= num:
    factorial *= i
    i += 1
print("Factorial:", factorial)
```
